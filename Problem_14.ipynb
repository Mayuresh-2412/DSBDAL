## Problem Statement 14: 
Locate dataset (e.g., sample_weather.txt) for working on weather data which reads the text input files 
and finds average for temperature, dew point and wind speed year wise/month-wise. 

```python
import pandas as pd
```

```python
df = pd.read_csv('Datasets/weather_data.csv')
df.describe(include='all')
```

```python
df['datetime'] = pd.to_datetime(df['date']+' '+df['time'])
```

```python
df['Year'] = df['datetime'].dt.year
df['Month'] = df['datetime'].dt.month
```

```python
df['temperature'] = pd.to_numeric(df['temperature'], errors='coerce')
df['dew_point'] = pd.to_numeric(df['dew_point'], errors='coerce')
df['wind_speed'] = pd.to_numeric(df['wind_speed'], errors='coerce')
```

```python
monthly_avg = df.groupby(['Year', 'Month'])[['temperature', 'dew_point', 'wind_speed']].mean().reset_index()
```
