## Problem Statement 11: 
Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for distribution of age 
with respect to each gender along with the information about whether they survived or not. (Column 
names: 'sex' and 'age') using boxplot and violinplot. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset("titanic")
titanic = titanic.dropna(subset=["age"])
sns.set(style="whitegrid")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
sns.boxplot(x='sex', y='age', hue='survived', data=titanic, ax=axes[0])
axes[0].set_title('Boxplot of AGe by Gender and Survival')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Age')
axes[0].legend(title='Survived', labels=['No', 'Yes'])
sns.violinplot(x='sex', y='age', hue='survived', data=titanic, ax=axes[1], split=True)
axes[1].set_title('Violinplot of Age by Gender and Survival')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Age')
axes[1].legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```
