## Problem Statement 12: 
Use the Iris flower dataset and perform the following operations. Scan the dataset and give the 
inference as:  
1. List down the features and their types (e.g., numeric, nominal) available in the dataset.  
2. Create a histogram for each feature in the dataset to illustrate the feature distributions.  
3. Create a boxplot for each feature in the dataset.  
4. Show the demonstration of the subplot. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris = sns.load_dataset("iris")
```

```python
print("\nFeatures")
print(iris.dtypes)
```

```python
numeric_columns = iris.select_dtypes(include='number').columns
iris[numeric_columns].hist(figsize=(10, 8), bins=20, edgecolor="black")
plt.suptitle("Histogram of Iris features")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,8))
for i,col in enumerate(numeric_columns, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=iris[col])
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 10))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(2, 2, i)
    plt.hist(iris[column], bins=20, edgecolor='black')
    plt.title(f'{column} Distribution')
    plt.xlabel(column)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.suptitle('Subplots of Iris Feature Distributions', y=1.02)
plt.show()
```
