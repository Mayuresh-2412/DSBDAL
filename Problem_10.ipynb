## Problem Statement 10: 
1. Use the inbuilt dataset 'titanic'.Use the Seaborn library to see if we can find any patterns in the data.  
2. Write a code to check how the price of the ticket (column name: 'fare') for each passenger is 
distributed by plotting a histogram, distplot,histplot,jointplot,rug plot,barplot. 


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
titanic = sns.load_dataset("titanic")
sns.set(style="whitegrid")
```

```python
# Histogram using histplot
plt.figure(figsize=(8, 6))
sns.histplot(data=titanic, x="fare", bins=30, kde=True, color="blue")
plt.title("HIstogram of Fare")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
# Distrubution using displot
plt.figure(figsize=(8, 6))
sns.displot(data=titanic, x="fare", bins=30, kde=True, color="blue", height=6, aspect=1.5)
plt.title("Distribution of Fare")
plt.xlabel("Fare")
plt.ylabel("Density")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=titanic, x="fare", bins=30, kde=True, color="green")
plt.title("Histogram of Fare with KDE")
plt.xlabel("Fare")
plt.ylabel("Density")
plt.show()
```

```python
# 4. Jointplot of Fare vs Age
plt.figure(figsize=(8, 6))
sns.jointplot(data=titanic, x="fare", y="age", kind="scatter", color="purple", height=6, ratio=5)
plt.suptitle("Jointplot of Fare vs Age", y=1.02)
plt.xlabel("Fare")
plt.ylabel("Age")
plt.show()
```

```python
# 5. Rugplot of fare
plt.figure(figsize=(2, 2))
sns.rugplot(data=titanic, x='fare', color='red')
plt.title("Rugplot of Fare")
plt.xlabel("Fare")
plt.ylabel("Density")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=titanic, x="class", y="fare", estimator='mean', palette="Blues", errorbar=None)
plt.title("Average Fare by Class")
plt.xlabel("Class")
plt.ylabel("Average Fare")
plt.show()
```
