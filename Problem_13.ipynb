## Problem Statement 13: 
Use the Iris flower dataset and perform the following operations. Scan the dataset and give the 
inference as:  
1. List down the features and their types (e.g., numeric, nominal) available in the dataset.  
2. Create a histogram for each feature in the dataset to illustrate the feature distributions.  
3. Compare distributions and identify outliers. 
4. Show the demonstration of the subplot. 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris = sns.load_dataset("iris")
```

```python
print("\nFeature Types:")
print(iris.dtypes)
print("\nFeature Names:")
print(iris.columns)
```

```python
iris.hist(figsize=(10, 8), edgecolor="black", bins=20)
plt.suptitle("Iris Dataset Histograms for features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for i, col in enumerate(iris.select_dtypes(include='number').columns):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x=iris[col])
    plt.title(f"Boxplot of {col}", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
for i, column in enumerate(iris.select_dtypes(include='number').columns):
    plt.subplot(2, 2, i + 1)
    plt.hist(iris[column], bins = 20, edgecolor='black')
    plt.title(f"{column} Distributions", fontsize=14)
    plt.xlabel(column)
    plt.ylabel("Frequency")
plt.tight_layout()
plt.suptitle("Iris Dataset Histograms for features", y=1.02, fontsize=16)
plt.show()
```
